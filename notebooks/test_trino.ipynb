```python
import os
import trino

user = os.environ.get("USER") or os.environ.get("NB_USER") or "unknown"

conn = trino.dbapi.connect(
    host="trino",
    port=8080,
    user=user,
    catalog="delta",
    schema="default",
)

cur = conn.cursor()
cur.execute("SHOW CATALOGS")
print(cur.fetchall())
```

```python
cur.execute("SHOW SCHEMAS FROM delta")
print(cur.fetchall())
```

```python
cur.execute("SHOW TABLES FROM kbase_ke_pangenome_2")
print(cur.fetchall())
```

```python
cur.execute("SELECT * FROM kbase_ke_pangenome_2.genome LIMIT 10")
rows = cur.fetchall()
rows
```

```python
cur.execute("DESCRIBE kbase_ke_pangenome_2.genome")
print(cur.fetchall())
```

```python
cur.execute("SELECT COUNT(*) FROM kbase_ke_pangenome_2.genome")
print(cur.fetchall())
```
